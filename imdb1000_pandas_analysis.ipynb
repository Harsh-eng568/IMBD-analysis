```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('/content/imdb-top-1000.csv')
movies
```

```python
genres = movies.groupby('Genre')
```

```python
# find top 3 genres by total earning
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre with highest average rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(5)
```

```python
# find the director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(5)
```

```python
# find the highest rated movie in each genre
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending=False).head(5)
```

```python
# find the number of movies done by each actor
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
movies.groupby('Genre').size()
```

```python
genre = movies.groupby('Genre')
genre.nth(6)
```

```python
genre.describe()
```

```python
  genre.agg(
      {
          'Runtime':'mean',
          'No_of_Votes':'sum',
          'Gross': 'sum',
          'IMDB_Rating': 'mean'
      }
  )
```

```python
df = pd.DataFrame(columns = movies.columns)
for group, data in genre:
  df = pd.concat([df, data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]], ignore_index=True)
df
```

```python
# find the number of movies starting with A for each group
def foo(group):
   group['Series_Title'].str.startswith('A')
   return group
```

```python
genre.apply(foo)
```

```python
# find ranking of each movie in the group according to IMBD score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genre.apply(rank_movie)
```

```python
# find normalized IMBD rating group wise
def normalize_rating(group):
  group['normalized_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
```

```python
genre.apply(normalize_rating)
```
